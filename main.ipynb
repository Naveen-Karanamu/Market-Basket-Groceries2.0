# ECLAT Implementaion

* Crating the instace which will replace the 1's with values and the rest with 0's

```python
eclat_instance = ECLAT(data=df, verbose=True)
```

```python
get_ECLAT_indexes, get_ECLAT_supports = eclat_instance.fit_all(min_support=0.2,
                                                           separator=' & ',
                                                           verbose=True)
```
